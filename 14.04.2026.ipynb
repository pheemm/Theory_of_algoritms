ri - коэффицент роста; n - количество месяцев; 1, 2, ... n - нумерация месяцев

r1 > r2 > r3 > ... > rn

Чем выше коэффицентр роста цены лицензии, тем быстрее необходимо ее приобрести.

Соответствено, если лицензия по номеру j покупается в месяце t, то получается ее стоимость:

100 * rj ** t

Значит, нам нужно оптимизировать порядок покупки лицензий и посчитать общую стоимость всех лицензий

Предположим, у нас есть две лицензии:

a обладает коэффицентом ra

b обладает коэффицентом rb

И ra > rb > 1

Вариант 1:

Купить a в месяц t, потом b в месяц t + 1 => C1 = 100(ra ** t + rb ** (t + 1))

Вариант 2:

Купить b в месяц t, потом a в месяц t + 1 => C2 = 100(rb ** t + rb ** (t + 1))

Теперь рассмотрим их разность:

C2 - C2 = 100(ra ** t + rb ** (t + 1)) - 100(rb ** t + rb ** (t + 1)) = 100((ra ** t)(ra - 1) - (rb ** t)(rb - 1))

Представим функцию при x = 1:

F(x) = x ** (t + 1) - x ** t

F'(x) = x ** (t - 1) * ((t + 1)x - t) > 0 => F(x) - строго возрастает

По предыдущим данным мы получаем вывод => C2 - C1 > 0 => C1 < C2

Если из двух лицензий одна растет быстрее, то ее нужно покупать раньше, чем другую

Следовательно любая последовательность, которая не соответствует: 

r1 > r2 > r3 > ... > rn

Не является оптимальной

17(2).

```python
from itertools import permutations

n = int(input())
r = list(map(float, input().split()))

best_money = -1
best_order = []

for order in permutations(range(n)):
    money = 0
    for month in range(n):
        item = order[month]
        money += 100 * (r[item] ** month)

    if money > best_money:
        best_money = money
        best_order = order

for item in best_order:
    print(item + 1, end=" ")
```
